```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os

print(os.getcwd())
```

```python
df = pd.read_csv("../data/Nestle_India_Financial_Data.csv")
```

```python
df
```

```python
df.info()
```

```python
# List of financial columns
financial_columns = [
    "Total Income (Rs. in million)",
    "Total Expenses (Rs. in million)",
    "Profit before Tax (Rs. in million)",
    "Profit After Tax (Rs. in million)"
]

# Remove commas and convert to numbers
for col in financial_columns:
    df[col] = df[col].astype(str).str.replace(",", "", regex=False)
    df[col] = pd.to_numeric(df[col])
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    df["Year"],
    df["Total Income (Rs. in million)"],
    marker='o',
    linewidth=3
)

plt.title("Nestle India Total Income Trend (2020-2026)")
plt.xlabel("Year")
plt.ylabel("Total Income (Rs. in million)")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    df["Year"],
    df["Total Expenses (Rs. in million)"],
    marker='o',
    markersize=8,
    linewidth=3,
    label="Total Expenses"
)

plt.title("Nestlé India Total Expenses Trend (2020–2026)", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Expenses (Rs. in million)", fontsize=12)

plt.xticks(df["Year"])
plt.legend()
plt.grid(True)

plt.tight_layout()

plt.savefig("../images/Total_Expenses_Trend.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    df["Year"],
    df["Profit before Tax (Rs. in million)"],
    marker='o',
    markersize=8,
    linewidth=3,
    label="Profit Before Tax"
)

plt.title("Nestlé India Profit Before Tax Trend (2020–2026)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Profit Before Tax (Rs. in million)")
plt.xticks(df["Year"])
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig("../images/Profit_Before_Tax_Trend.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    df["Year"],
    df["Profit After Tax (Rs. in million)"],
    marker='o',
    markersize=8,
    linewidth=3,
    label="Profit After Tax"
)

plt.title("Nestlé India Profit After Tax Trend (2020–2026)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Profit After Tax (Rs. in million)")
plt.xticks(df["Year"])
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig("../images/Profit_After_Tax_Trend.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df["Year"],
         df["Total Income (Rs. in million)"],
         marker='o',
         linewidth=3,
         label="Total Income")

plt.plot(df["Year"],
         df["Total Expenses (Rs. in million)"],
         marker='s',
         linewidth=3,
         label="Total Expenses")

plt.title("Nestlé India Income vs Expenses")
plt.xlabel("Year")
plt.ylabel("Rs. in million")

plt.xticks(df["Year"])
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig("../images/Income_vs_Expenses.png", dpi=300)

plt.show()
```

```python
df["Income Growth (%)"] = df["Total Income (Rs. in million)"].pct_change()*100

df["Expense Growth (%)"] = df["Total Expenses (Rs. in million)"].pct_change()*100

df["PAT Growth (%)"] = df["Profit After Tax (Rs. in million)"].pct_change()*100

df.round(2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Independent Variable (Year)
X = df[["Year"]]

# Dependent Variable (Total Income)
y = df["Total Income (Rs. in million)"]
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
future_years = pd.DataFrame({
    "Year":[2027,2028,2029]
})
```

```python
predicted_income = model.predict(future_years)
```

```python
forecast = future_years.copy()

forecast["Predicted Total Income"] = predicted_income

forecast
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

# Historical Data
plt.plot(df["Year"], df["Total Income (Rs. in million)"], marker='o', label='Historical Total Income')

# Forecasted Data
plt.plot(forecast["Year"], forecast["Predicted Total Income"], marker='o', linestyle='--', label='Forecasted Total Income')

plt.title("Historical vs Forecasted Total Income")
plt.xlabel("Year")
plt.ylabel("Total Income (Rs. in million)")
plt.legend()
plt.grid(True)
plt.show()
```
